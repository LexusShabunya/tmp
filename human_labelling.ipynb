# Labelling

```python
def label(text: str, class_map: dict):
    class_ = input('Метка класса: ')
    while not class_.isdigit():
        class_ = input('Метка класса: ')
    
    class_ = int(class_)
    if class_ in class_map:
        class_name = class_map[class_]
    else:
        class_name = input('Название класса: ')
        
    return class_, class_name
```

```python
with open(f'classes.npy', 'rb') as file:
    classes = np.load(file).tolist()
with open(f'class_map.txt', 'r', encoding='utf-8') as file:
    for line in file:
        key, value = line.rstrip().split(': ')
        class_map[int(key)] = value
```

```python
cnt = len(classes)
for text in tqdm(data['text'].tolist()[len(classes):]):
                
    print(class_map)
    print()
    print(text)
    
    class_, class_name = label(text, class_map)
    class_map[class_] = class_name
    
    classes.append(class_)
    cnt += 1
    
    print(120*'-')
```
